```python
values=[5,6,3,7,2]
names=["A","B","C","D","E"]

plt.barh(names, values, color=["green","blue","red","grey","black"])
plt.show()
```

```python
ages=[84,74,13,18,36,32,95,59,4,98,7,5,49,61,93,81,35,77,87,61,12,74,37,44,40,14,1,84,14,74,62,51,88,76,73,27,25,24,69,1]
range=(0,100)
bins=10
plt.hist(ages, bins, range, color="green", histtype='bar',edgecolor='black', alpha=0.23)

plt.title("My Histrogram")
plt.xlabel("Ages")
plt.ylabel("No. of People")

plt.show()
```

```python
x_valss=[0,1,2,3,4,5]
y_valss=[0,1,4,9,16,25]

plt.scatter(x_valss, y_valss, s=50, color="blue",marker="*")
plt.show()
```

```python
activities=['sleeping','eating','working','playing']
slices=[7,2,2,3]
colors=['r','g','b','y']

plt.pie(slices, labels=activities, colors=colors, startangle=90, shadow=True, radius=2, explode=(0,0.1,0,0), autopct='%1.1f%%')
plt.show()
```
